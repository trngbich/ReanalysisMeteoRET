```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import itertools
import matplotlib.gridspec as gridspec
```

# 2018-2022 difference plot
## t_air, vp, u

```python
csv_path=r'./data/validation_{0}.csv'
product1='GEOS5'
product2='AgERA5'
metrics = ['bias','nse','r2','rmse']
var_names = ['t_air',
             # 'p_air',
             # 'ra',
             'u','vp']
data_dict={
'variable': [],
'N': [],
# 'station':[],
 'bias': [],
'nse': [],
'r2': [],
'rmse': []
}
```

```python
for var_name in var_names:
    csv=csv_path.format(var_name)
    df= pd.read_csv(csv,sep=';')        
    # data_dict['station']+=list(df['Station'].values)
    data_dict['N']+=list(df['{0}_GEOS5_N'.format(var_name)].values)
    data_dict['variable']+=list(itertools.repeat(var_name, len(df)))
    for metric in metrics:        
        col1='{0}_{1}_{2}'.format(var_name, product1, metric)
        col2='{0}_{1}_{2}'.format(var_name, product2, metric)        
        diff=df[col2]-df[col1]
        data_dict[metric]+=list(diff)                            
df_all=pd.DataFrame(data_dict)
df_all=df_all.dropna()                                
        
        
        
```

```python
plt.figure(figsize=(7, 5))
ax=sns.stripplot(x='variable', 
              y='r2',               
              hue='N', #replace with land cover class
              data=df_all,      
              # order=y_order, #reorder y axis
              hue_order=['1','>1 to 10', '>10 to 100', '>100','NA'],
              # palette=['#648FFF', '#785EF0','#DC267F','#FE6100','black'], 
              alpha=0.5, size=5)
ax.set_xlabel('Variables')
ax.set_ylabel('AgERA5 - GEOS5')
ax.set_title('r2 difference')
ax.set_ylim([-0.6,0.6])
ax.axhline(y=0,color='black')
ax.axhspan(0,0.6,facecolor='green',
                alpha=0.2,zorder=-1000)
ax.axhspan(-0.6,0,facecolor='red',
                alpha=0.2,zorder=-1000)
plt.legend(ncol=2,title='Number of data points')
```

# Scatterplot

```python
def plot_metrics_scatter(x, y,z, xlabel='X',ylabel='Y',title = '',metric='r2',ax=None,show_cbar=False):
    if ax is None:
        ax = plt.gca()
    scatter=ax.scatter(x,y,c=z,s=10,
                       # edgecolors='black',linewidth =1 
                      )
        
#     # Create scatter plot for points above the 1:1 line
#     ax.scatter(x[y >= x], y[y >= x], color='green', label='Above 1:1 line')

#     # Create scatter plot for points below the 1:1 line
#     ax.scatter(x[y < x], y[y < x], color='red', label='Below 1:1 line')

    # Plot 1:1 line
    # ax.plot([vmin, vmax], [vmin, vmax], color='black', linestyle='--')
    
    vmin=min(np.nanmin(metric1),np.nanmin(metric2))
    vmax=max(np.nanmax(metric1),np.nanmax(metric2))    
### evaluation graph
    if metric=='r2':
        ax.set_xlim(0, 1)
        ax.set_ylim(0, 1)        
        # Create polygon for the area above the 1:1 line
        polygon_points = np.array([[1, 0], [0, 0], [1, 1]])
        ax.fill(polygon_points[:, 0], polygon_points[:, 1], color='red', alpha=0.1,zorder=-10000)

        # Create polygon for the area below the 1:1 line
        polygon_points = np.array([[0, 1], [0, 0], [1, 1]])
        ax.fill(polygon_points[:, 0], polygon_points[:, 1], color='green', alpha=0.1,zorder=-10000)    
        
    if metric=='rmse':
        ax.set_xlim(vmin, vmax)
        ax.set_ylim(vmin, vmax)        
        # Create polygon for the area above the 1:1 line
        polygon_points = np.array([[vmax, vmin], [vmin, vmin], [vmax, vmax]])
        ax.fill(polygon_points[:, 0], polygon_points[:, 1], color='green', alpha=0.1,zorder=-10000)

        # Create polygon for the area below the 1:1 line
        polygon_points = np.array([[vmin, vmax], [vmin, vmin], [vmax, vmax]])
        ax.fill(polygon_points[:, 0], polygon_points[:, 1], color='red', alpha=0.1,zorder=-10000)  
        
    if metric=='bias':
        ax.set_xlim(vmin, vmax)
        ax.set_ylim(vmin, vmax)        
        vmax=max(abs(vmax),abs(vmin))
        # y better than x area
        polygon_points = np.array([[0, 0], [vmax, vmax], [vmax, -vmax]])
        ax.fill(polygon_points[:, 0], polygon_points[:, 1], color='green', alpha=0.1,zorder=-10000)
        polygon_points = np.array([[0, 0], [-vmax, -vmax], [-vmax, vmax]])
        ax.fill(polygon_points[:, 0], polygon_points[:, 1], color='green', alpha=0.1,zorder=-10000)        

        # x better than y area
        polygon_points = np.array([[0, 0], [vmax, vmax], [-vmax, vmax]])
        ax.fill(polygon_points[:, 0], polygon_points[:, 1], color='red', alpha=0.1,zorder=-10000)
        polygon_points = np.array([[0, 0], [-vmax, -vmax], [vmax, -vmax]])
        ax.fill(polygon_points[:, 0], polygon_points[:, 1], color='red', alpha=0.1,zorder=-10000)   

    if metric=='nse':
        ax.set_xlim(vmin, 1)
        ax.set_ylim(vmin, 1)        
        vmax=max(abs(vmax),abs(vmin))
        # Create polygon for the area above the 1:1 line
        polygon_points = np.array([[1, vmin], [vmin, vmin], [1, 1]])
        ax.fill(polygon_points[:, 0], polygon_points[:, 1], color='red', alpha=0.1,zorder=-10000)

        # Create polygon for the area below the 1:1 line
        polygon_points = np.array([[vmin, 1], [vmin, vmin], [1, 1]])
        ax.fill(polygon_points[:, 0], polygon_points[:, 1], color='green', alpha=0.1,zorder=-10000)        
         
    
    # Add colorbar
    if show_cbar:
        cbar = plt.colorbar(scatter)
        cbar.set_label('Number of data points')    

    # Set x-axis and y-axis equal
    ax.set_aspect('equal')    
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_title(title)
    # ax.legend()
    # ax.grid(True)
    return scatter

```

```python
csv_path=r'./data/validation_{0}.csv'
product1='GEOS5'
product2='AgERA5'
# metrics = ['nse']
metrics = {'bias':'Bias',
           'nse':'NSE',
           'r2':'r$^2$',
           'rmse':'RMSE'}
var_names = {'t_air':'Air temperature (Celcius degree)',
             # 'p_air',
             # 'ra',
             'u':'Windpseed (m/s)',
             'vp':'Vapor pressure (mbar)'
            }
data_dict={
'variable': [],
'N': [],
# 'station':[],
 'bias': [],
'nse': [],
'r2': [],
'rmse': []
}
```

```python
for var_name in var_names:
    csv=csv_path.format(var_name)
    df= pd.read_csv(csv,sep=';')        
    # data_dict['station']+=list(df['Station'].values)
    data_dict['N']+=list(df['{0}_GEOS5_N'.format(var_name)].values)
    data_dict['variable']+=list(itertools.repeat(var_name, len(df)))
    n_points=df['{0}_GEOS5_N'.format(var_name)]
    # fig,axs = plt.subplots(1,5,figsize=(20,5))
    fig = plt.figure(figsize =(15,5))
    gs = gridspec.GridSpec(2, 41) #legend is 1:10 of plot
    axs=[plt.subplot(gs[0,:10]),
         plt.subplot(gs[0,10:20]),
         plt.subplot(gs[0,20:30]),
         plt.subplot(gs[0,30:40]),
         plt.subplot(gs[0,40:41])
        ]
    gs.update(wspace = 0.1, hspace = 0.1)    
    for i,metric in enumerate(metrics):        
        col1='{0}_{1}_{2}'.format(var_name, product1, metric)
        col2='{0}_{1}_{2}'.format(var_name, product2, metric)        
        metric1=df[col1]
        metric2=df[col2]
        scatter=plot_metrics_scatter(metric1,metric2,n_points,
                             metric=metric,
                             xlabel=product1, ylabel=product2, 
                             title=metrics[metric],
                             ax=axs[i],
                            )    
    
    cbar = fig.colorbar(scatter,cax=axs[4])
    cbar.set_label('Number of data points')
    tick_max=int(round(int(np.nanmax(n_points))/100)*100)
    cbar.set_ticks(np.linspace(0, tick_max, 5))
    cbar.set_ticklabels(['{:.0f}'.format(val) for val in np.linspace(0, tick_max, 5)])
    fig.suptitle(var_names[var_name])
    plt.show()
```
