```python
import numpy as np
import pandas as pd
import os
import glob
```

# all stations average

```python
variables = {'t_air':['GEOS5','AgERA5'],
             'p_air':['GEOS5','ERA5'],
             'u':['GEOS5','AgERA5'],
             'vp':['GEOS5','AgERA5'],
             'ra':['MSG','GEOS5']
            }
metrics = ['bias','nse','r2','rmse']

path = r'./data/validation_{0}.csv'

data = {
    'N_stations': [],
    'variable': []
}

for metric in metrics:
    data[f'v2_{metric}']=[]
    data[f'v3_{metric}']=[] 
    data[f'v3-v2_{metric}']=[]
for variable in variables:
    data['variable'].append(variable)
    fh = path.format(variable)
    df = pd.read_csv(fh, sep=';')
    data['N_stations']=df[f'{variable}_GEOS5_N'].count()
    for metric in metrics:
        v2_metric=df[f'{variable}_{variables[variable][0]}_{metric}'].mean()
        v3_metric=df[f'{variable}_{variables[variable][1]}_{metric}'].mean()
        data[f'v2_{metric}'].append(v2_metric)
        data[f'v3_{metric}'].append(v3_metric)
        data[f'v3-v2_{metric}'].append(v3_metric-v2_metric)
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.to_csv(r'./data/all_stations_avg_metrics.csv',sep=';')
```

# all station v2, v3 difference

```python
variables = {'t_air':['GEOS5','AgERA5'],
             'p_air':['GEOS5','ERA5'],
             'u':['GEOS5','AgERA5'],
             'vp':['GEOS5','AgERA5'],
             'ra':['MSG','GEOS5']
            }
path = r'./data/validation_{0}_split.csv'

data = {    
    'variable': [],
    'v3-v2_MD':[],
    'v3-v2_MD_before':[],
    'v3-v2_MD_after':[],
}

for variable in variables:
    data['variable'].append(variable)
    fh = path.format(variable)
    df = pd.read_csv(fh, sep=';')      
    v2_val=df[f'{variables[variable][0]}_mean'].mean()
    v3_val=df[f'{variables[variable][1]}_mean'].mean()    
    v2_before=df[f'{variables[variable][0]}_mean_before'].mean()
    v3_before=df[f'{variables[variable][1]}_mean_before'].mean()  
    v2_after=df[f'{variables[variable][0]}_mean_after'].mean()
    v3_after=df[f'{variables[variable][1]}_mean_after'].mean()      
    data['v3-v2_MD'].append(v3_val-v2_val)
    data['v3-v2_MD_before'].append(v3_before-v2_before)
    data['v3-v2_MD_after'].append(v3_after-v2_after)    
```

```python
df = pd.DataFrame(data)
```

```python
df.to_csv(r'./data/all_stations_v3-v2_MD.csv',sep=';')
```
