```python
import pandas as pd
import glob
import os
```

# Tanzania

```python
folder=r"D:\SurfDrive\Projects\WaPOR phase 2\3_Deliverables\2_Quality Assessment\0_Insitu_Data\TAHMO\Data_hourly_Tanzania"
fhs=glob.glob(os.path.join(folder,'*.csv'))
```

## Check fh with missing columns

```python
vars_list = ['atmosphericpressure AVG (kPa)', #p_air
             'radiation AVG (W/m2)', #ra
             'temperature AVG (degrees Celsius)', #t_air
             'temperature MIN (degrees Celsius)',
             'temperature MAX (degrees Celsius)',
             'windspeed AVG (m/s)', #u
             'relativehumidity AVG (-)', #rh
            ]
fhs_missing = []
for fh in fhs:
    print(os.path.basename(fh))
    df=pd.read_csv(fh,parse_dates=True,index_col=0)
    miss_col = False
    for var in vars_list:
        if var not in df.columns:            
            print('Missing '+var)
            miss_col=True
    if miss_col is True:
        fhs_missing.append(fh)
```

```python
fhs_missing
```

## Merge sensor columns

```python
vars_list = {'atmosphericpressure':'atmosphericpressure {0} (kPa)', #p_air
             'radiation':'radiation {0} (W/m2)', #ra
             'temperature':'temperature {0} (degrees Celsius)', #t_air
             'windspeed':'windspeed {0} (m/s)', #u
             'relativehumidity':'relativehumidity {0} (-)', #rh
            }
sensors_meta=pd.read_csv(fhs_missing[0])
save_folder=r"D:\SurfDrive\Projects\WaPOR phase 2\3_Deliverables\2_Quality Assessment\0_Insitu_Data\TAHMO\Data_hourly_Tanzania\Merge_sensors"
for fh in fhs_missing[2:]:
    # fh=fhs_missing[3]
    df=pd.read_csv(fh, index_col=0, parse_dates=True)
    print(os.path.basename(fh))
    for var in vars_list:        
        for stat in ['AVG','MIN','MAX']:
            print(var+' '+stat)
            # get list of column that has variable
            var_columns=[column for column in df.columns if ((var+" S" in column) and (stat in column))]

            for column in var_columns:            
                sensor_code = column.split(' ')[1]
                sensor_info = sensors_meta.loc[sensors_meta['sensor code']==sensor_code]
                sensor_name = sensor_info.values[0][1] #get sensor name
                if sensor_name != 'ATMOS 41': #check if column use sensor is not ATMOS41 
                    var_columns.remove(column)            
                print('{0} {1}'.format(sensor_code,sensor_name))
            #combine all columns in var_columns into a new column in df
            if len(var_columns)>0:
                df_columns = [df[column] for column in var_columns]    
                new_colmn = vars_list[var].format(stat) #get name of variable column
                # if new_colmn == "temperature MAX (degrees Celsius)":
                #     print("WORKED")
                df[new_colmn]=pd.concat(df_columns, axis=1).mean(axis=1)
    #save df to new csv
    save_filename = os.path.basename(fh)
    save_fh = os.path.join(save_folder,save_filename)
    df.to_csv(save_fh,sep=';')
```
