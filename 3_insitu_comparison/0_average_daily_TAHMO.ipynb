```python
skipped_fhs=[]
for hour_fh in tqdm(hour_fhs):
    try:
        filename=os.path.basename(hour_fh)
        df_hour = pd.read_csv(hour_fh,parse_dates=True,index_col=0,sep=";|,")
        ### RESAMPLE TO DAILY
        # Define columns for which you want to change the aggregation method
        columns_to_change = ['temperature MAX (degrees Celsius)','temperature MIN (degrees Celsius)']  # Replace 'column1', 'column2', etc. with your actual column names

        # Create a dictionary mapping columns to their respective aggregation methods
        aggregation_methods = {'temperature MIN (degrees Celsius)': 'min',  # Change 'min' to 'max' if you want maximum
                               'temperature MAX (degrees Celsius)': 'max'}  # Change 'min' to 'max' if you want maximum
        # Apply different aggregation methods to specified columns
        df_day_agg = df_hour.resample('D').agg(aggregation_methods)

        # If you want to keep the same aggregation method for other columns, you can still use 'mean' for them
        df_day = df_hour.resample('D').mean()
        df_day[columns_to_change] = df_day_agg[columns_to_change]
        
        #### DONE####
        save_fh = os.path.join(save_folder,filename)
        df_day.to_csv(save_fh)
    except:
        # print('Skipped '+hour_fh)
        skipped_fhs.append(hour_fh)
        continue
    
```
